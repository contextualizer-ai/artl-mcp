```python
from artl_mcp.tools import search_papers_by_keyword
```

```python
# Search for hydrocarbon microbiome papers, 50 results, sorted by citation count
results = search_papers_by_keyword(
  query="hydrocarbon microbiome",
  max_results=10,
  sort="relevance"
)
```

```python


# Access the results
if results:
  papers = results["message"]["items"]
  print(f"Found {len(papers)} papers")

  # Show first paper details
  if papers:
      first_paper = papers[0]
      print(f"Title: {first_paper.get('title', ['Unknown'])[0]}")
      print(f"DOI: {first_paper.get('DOI', 'Unknown')}")
      print(f"Citation count: {first_paper.get('is-referenced-by-count', 0)}")
```
